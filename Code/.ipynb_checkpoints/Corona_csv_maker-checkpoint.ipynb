```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
sel_mun = ["Albertslund","Ballerup","Brøndby","Dragør","Frederiksberg","Gentofte","Gladsaxe",
                           "Glostrup","Herlev","Hvidovre","Høje-Taastrup","Hørsholm","Ishøj","København",
                           "Rødovre","Taarnby"]
```

```python
df = pd.read_csv("Covid-19_indlæggelser_1804.csv",sep=";").dropna() #encoding="ISO8859-1"

df = df.set_index("Kommune")

df
```

```python
df_pct = df.div(df.loc["Hovedtotal"],1)*100

df_pct
```

```python
fig = plt.figure(figsize=(20,20))

for k,num in zip(df.index, range(1,len(df.index))):
    #df0=df[df['Country']==c]
    #df0 = df_pct.loc[k]
    ax = fig.add_subplot(10,3,num,)
    ax.plot(df_pct.columns, df_pct.loc[k])
    ax.set_title(k,)
    ax.set_xticklabels(df.columns,rotation=45)
    ax.set_ylim

plt.tight_layout()
plt.show()
```

```python
df_pct.iloc[:,0:15]
```

```python
pre_host = np.array([[26,36,64,77,90,104,20,131,150,170,193,215,230,240,259,258,261]])

dates = np.array(["16.mar","17.mar","18.mar","19.mar","20.mar","21.mar","22.mar","23.mar","24.mar","25.mar",
                  "26.mar","27.mar","28.mar","29.mar","30.mar","31.mar","1.apr"])

df_pct['avg_pct'] = df_pct.iloc[:,0:15].mean(1)

avg_pct = np.array([df_pct['avg_pct'].iloc[:-1]])/100
```

```python
pre_df = pd.DataFrame(np.dot(avg_pct.T,pre_host),index=df.index[:-1],columns=dates).apply(np.round)
```

```python
pre_df.loc["Hovedtotal"] = pre_df.sum(0)
```
