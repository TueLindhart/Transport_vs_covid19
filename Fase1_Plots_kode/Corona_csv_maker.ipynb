```python
pre_host = np.array([[26,36,64,77,90,104,120,131,150,170,193,215,230,240,259,258,261]])

dates = np.array(["16.mar","17.mar","18.mar","19.mar","20.mar","21.mar","22.mar","23.mar","24.mar","25.mar",
                  "26.mar","27.mar","28.mar","29.mar","30.mar","31.mar","01.apr"])

df_pct.loc['avg_pct'] = df_pct.mean(0)

avg_pct = np.array([df_pct.loc['avg_pct'].iloc[:-1]])/100
```

```python
pre_df_unrounded = pd.DataFrame(np.dot(pre_host.T,avg_pct),index=dates,columns=df.columns[:29])

pre_df = pre_df_unrounded.apply(np.round)
diff_df = pre_df_unrounded - pre_df

temp_total = np.array([pre_df.iloc[:,:29].sum(1)])

pre_diff = pre_host - temp_total

pre_diff = [int(i) for i in pre_diff.squeeze()]

row = 0
idx = list(range(len(pre_diff)))

#Removes hospitalisations closest to not becoming rounded up if the sum per date exceeds pre_host total
# Adds hospitalisations closest to becoming rounded up if the sum per data is below pre_host total 
for row,i in enumerate(pre_diff):
    temp = diff_df.iloc[row,:]
    
    if i > 0:
        max_c = temp.nlargest(i).index.values.tolist()
        for c in max_c:
            pre_df.loc[:,c][row] = pre_df.loc[:,c][row] + 1
    if i < 0:
        min_c = temp.nsmallest(-i).index.values.tolist()
        for c in min_c:
            pre_df.loc[:,c][row] = pre_df.loc[:,c][row] - 1
```

```python
pre_df["Hovedtotal"] = pre_df.iloc[:,:29].sum(1)
```

```python
[pre_df["Hovedtotal"]] - pre_host
```

```python
post_df = np.array([246,240,228,231,237,218,213,201,199,199,194,192,196,188,183,184,163,167,178,182,177])

mean_post_df = np.mean(post_df)

mean_post_muns = df["Hovedtotal"].mean()

hosp_in_cph = mean_post_muns / mean_post_df

pre_df = (pre_df * hosp_in_cph).round()

pre_df["Hovedtotal"] = pre_df.iloc[:,:29].sum(1)
```

```python
df.loc["17.apr"] = df.loc["16.apr"] + ((df.loc["19.apr"] - df.loc["16.apr"])/2).round()
df.loc["18.apr"] = df.loc["17.apr"] + (df.loc["19.apr"] - df.loc["17.apr"])

df = df.reset_index().sort_values(by="index").set_index("index")
```

```python
full_df = pd.concat([pre_df,df],sort=False)
```
